```python
x_cat = df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
```

```python
x_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
x_cat = onehotencoder.fit_transform(x_cat).toarray()
```

```python
x_cat.shape
```

```python
x_cat = pd.DataFrame(x_cat)
```

```python
x_cat
```

```python
x_numerical = df.drop(columns = {'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus','Attrition'})
```

```python
x_all = pd.concat([x_cat, x_numerical], axis = 1)
```

```python
x_all.info()
```

```python
x_all.columns = x_all.columns.astype(str)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x_all)
```

```python
x
```

```python
df['Attrition'] = df['Attrition'].astype(int)
```

```python
y = df['Attrition']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.25)
```

```python
X_train.shape
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print('Accuracy {} '.format(100 * accuracy_score(y_pred, Y_test)))
```

```python
cm = confusion_matrix(y_pred, Y_test)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(Y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_pred, Y_test)
```

```python
cm
```

```python
print(classification_report(Y_test, y_pred))
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu', input_shape = (76, )))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
epochs_hist = model.fit(X_train, Y_train, epochs = 100, batch_size = 50)
```

```python
Y_pred = model.predict(X_test)
Y_pred = (Y_pred > 0.5)
```

```python
plt.plot(epochs_hist.history['loss'])
```

```python
cm = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cm, annot = True )
```

```python
print(classification_report(Y_test, Y_pred))
```
